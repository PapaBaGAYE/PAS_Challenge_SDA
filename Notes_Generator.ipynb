```python
import numpy as np
import pandas as pd
from math import sqrt
```

```python
mu = 12
sigma = 15
```

```python
def genererReleves_S1(n):
  np.random.seed(1)
  list_releves = np.random.randn(n, 7) * sqrt(sigma) + mu
  list_releves = np.round(list_releves)
  list_releves[list_releves < 6] = 6
  list_releves[list_releves > 19] = 19
  df = pd.DataFrame(data=list_releves, columns=[i for i in ["Maths", "PC", "SVT", "Philo", "Fr", "Anglais", "HG"]])
  df.to_csv(f"releve_de_{n}_etudiants_series_S1.csv", index=False)
  return df
  
genererReleves_S1(1000)
```

```python
def genererReleves_S2(n):
  np.random.seed(2)
  list_releves = np.random.randn(n, 7) * sqrt(sigma) + mu
  list_releves = np.round(list_releves)
  list_releves[list_releves < 6] = 6
  list_releves[list_releves > 19] = 19
  df = pd.DataFrame(data=list_releves, columns=[i for i in ["Maths", "PC", "SVT", "Philo", "Fr", "Anglais", "HG"]])
  df.to_csv(f"releve_de_{n}_etudiants_series_S2.csv", index=False)
  return df

genererReleves_S2(1000)
```

```python
def genererReleves_L1(n):
  np.random.seed(3)
  list_releves = np.random.randn(n, 6) * sqrt(sigma) + mu
  list_releves = np.round(list_releves)
  list_releves[list_releves < 6] = 6
  list_releves[list_releves > 19] = 19
  df = pd.DataFrame(data=list_releves, columns=[i for i in ["Philo", "HG", "Fr", "Anglais", "LV2", "Maths"]])
  df.to_csv(f"releve_de_{n}_etudiants_series_L1.csv", index=False)
  return df

genererReleves_L1(1000)
```

```python
def genererReleves_L2(n):
  np.random.seed(4)
  list_releves = np.random.randn(n, 7) * sqrt(sigma) + mu
  list_releves = np.round(list_releves)
  list_releves[list_releves < 6] = 6
  list_releves[list_releves > 19] = 19
  df = pd.DataFrame(data=list_releves, columns=[i for i in ["Philo", "HG", "Fr", "Anglais", "LV2", "PC_SVT", "Maths"]])
  df.to_csv(f"releve_de_{n}_etudiants_series_L2.csv", index=False)
  return df

genererReleves_L2(1000)
```

```python
data = genererReleves_S2(1000)
data["Moy"] = round((data.Maths*5 + data.PC*6 +	data.SVT*6 + data.Philo*2	+ data.Fr*3	+ data.Anglais*2	+ data.HG*2) / 26, 2)
data.head()
```
